```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import sklearn
%matplotlib inline
```

## 练习1,2

为了方便测试，对knn2的代码作了修改

```python
import knn2
```

```python
error_rate_l1 = []
error_rate_l2 = []
for k in range(1, 21):
    error_rate_l1.append(knn2.datingClassTest(k,'l1'))
    error_rate_l2.append(knn2.datingClassTest(k,'l2'))
```

使用不同的k值和使用L1和L2距离对错误率的影响如下图

```python
plt.plot(error_rate_l1, label='l1')
plt.plot(error_rate_l2, label='l2')
plt.legend()
```

结论：
- 过小或过大的k值都会影响预测的结果
- 使用L2距离的准确率要高于使用L1距离

## 练习3

```python
sizes = []
error_rate = []
for i in range(1,21):
    size = i / 20
    sizes.append(size)
    error_rate.append(knn2.datingClassTest(6,'l2', size))
```

不同样本数目对错误率的影响如下图

```python
plt.plot(sizes, error_rate)
```

结论：
- 总体来说，样本数目越多，准确率越高
